```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open('data/E-GEOD-61162/E-GEOD-61162-analytics.tsv')
```

```python
data = []
for line in file:
    line =  line.split()
    data.append(line)
    
```

```python
data[2][3]
```

```python
log2fc =  np.zeros(0)
for i in range(1,len(data)-10):
    #if len(data[i]) == 4:
        
    log2fc = np.append(log2fc, float(data[i][-1]))
    
```

```python
len(log2fc)
```

```python
plt.hist(log2fc)
```

```python
idx  = (log2fc>=1.8)
names = []

for i in range(1,len(idx)):
    if idx[i] == True:
        names.append(data[i][1])
```

```python
f = open('protein_name.txt', 'a')
for name in names:
    f.write(name)
    f.write("\n")
    
```
